# Gas + laser 45% — dwell comparison

Overlay all `*GAS_LASER_ON_45%*` spectra on one plot; each integration time gets its own color and legend entry.

**Kernel:** Python (TOF Analysis)

```python
%matplotlib inline

from pathlib import Path

import matplotlib.pyplot as plt
from tof_analysis.plotting import (
    default_dwell_overlay_ions,
    find_spectra_by_label_pattern,
    format_dwell_label,
    plot_dwell_series,
)

ROOT = Path("..").resolve()
DATA = ROOT / "DATA"
CAL = ROOT / "calibration"
FIGURES = ROOT / "figures"
FIGURES.mkdir(exist_ok=True)

PATTERN = "*GAS_LASER_ON_45%*"

pairs = find_spectra_by_label_pattern(DATA, PATTERN, gas=True)
print(f"Matched {len(pairs)} GAS files:")
for meta, _ in pairs:
    print(f"  {format_dwell_label(meta):>8}  {meta.label}")

ions = default_dwell_overlay_ions()
print(f"\nOverlay: {len(ions)} ion lines (Ar, Ho, CH4 fragments q=1, H+, D+)")

ax = plot_dwell_series(
    DATA,
    PATTERN,
    gas=True,
    calibration=CAL / "reference.json",
    overlay_ions=ions,
    title="GAS + laser 45% — integration time comparison",
    save_path=FIGURES / "gas_laser_45_dwell_series.png",
    show=True,
)
plt.show()
print(f"Saved {FIGURES / 'gas_laser_45_dwell_series.png'}")
```
